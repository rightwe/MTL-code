```python
import pandas as pd

original_data = pd.read_csv('./xenonpy_input.csv')
original_data['Mn'] = original_data['Mn'] + original_data['Mn_2']
original_data['Sn'] = original_data['Sn'] + original_data['Sn_2']
original_data.drop(columns=['Mn_2', 'Sn_2'], inplace=True)
original_data
```

```python
data = original_data.div(original_data.sum(axis=1), axis=0)*100
data
```

```python
def generat_formula(x):
    formula = ''
    for key, value in x.items():
        if key == 'formula':
            continue
        if value > 0:
            formula += f"{key}{round(value, 3)}"
    return formula
original_data['formula'] = data.apply(generat_formula, axis = 1)
```

```python
original_data.to_csv('./xenonpy_input.csv', index=False)
```
